# <center> CONEXION A LA BASE DE DATOS</center>

```python
# https://docs.sqlalchemy.org/en/14/core/engines.html#database-urls
# PostgreSQL, MySQL, Oracle, SQL Server, SQLite...
# https://github.com/googleapis/python-bigquery-sqlalchemy

#!pip install sqlalchemy
#!pip install cryptography
#!pip install pymysql
#!pip install psycopg2
#!pip install pandas

from sqlalchemy import create_engine


def connection_string(engine:str = 'mysql'):

    database = 'digitos'
    user = 'admin'
    password = 12345678
    host = 'localhost'
    
    engines = {
        'mysql':{
            'engine':'mysql+pymysql',
            'port': 3333
            },
        'postgres':{
            'engine':'postgresql+psycopg2',
            'port': 5555
            },
        'sqlserver':{
            'engine':'mssql+pymssql',
            'port': 1433
            },
        'oracle':{
            'engine':'oracle',
            'port': 1521
            }     
    }
    
    values = engines.get(engine)
    engine = values.get('engine','mysql')
    port = values.get('port','3333')
        
    return f'{engine}://{user}:{password}@{host}:{port}/{database}'
```

```python
from pprint import pprint as print

print('MySQL CONNECTION')
mysql = create_engine(connection_string('mysql'))
print(mysql.__dict__)

print('-----------------------------------------------')

print('PostgreSQL CONNECTION')
postgres = create_engine(connection_string('postgres'))
print(postgres.__dict__)
```

# <center> CREAR LAS TABLAS DE LA BASE DE DATOS </center>

```python
query = """

CREATE TABLE IF NOT EXISTS numeros (
   nombre_numero VARCHAR(255) NOT NULL,
   valor INT NOT NULL,
   cifras INT NOT NULL,
   creado TIMESTAMP NOT NULL,
   PRIMARY KEY (nombre_numero)
);


CREATE TABLE IF NOT EXISTS fibonacci (
   nombre_numero VARCHAR(255) NOT NULL,
   numero_fibonacci VARCHAR(255) NOT NULL,
   cifras INT NOT NULL,
   PRIMARY KEY (nombre_numero)
);

CREATE TABLE IF NOT EXISTS par (
   nombre_numero VARCHAR(255) NOT NULL,
   es_par INT NOT NULL,
   PRIMARY KEY (nombre_numero)
);

CREATE TABLE IF NOT EXISTS divisible_por_5 (
   nombre_numero VARCHAR(255) NOT NULL,
   es_divisible INT NOT NULL,
   PRIMARY KEY (nombre_numero)
);

CREATE TABLE IF NOT EXISTS colores (
   idcolores INT NOT NULL,
   precio INT DEFAULT NULL,
   descripcion VARCHAR(255) NOT NULL,
   PRIMARY KEY (idcolores)
);

CREATE TABLE colores_auditoria (
   idcolores INT  NULL,
   precio INT  NULL,
   descripcion VARCHAR(255)  NULL,
   accion  VARCHAR(255)  NULL,
   usuario VARCHAR(255)  NULL,
   entry_date TEXT  NULL
);

CREATE TABLE IF NOT EXISTS universo (
   iduniverso INT NOT NULL,
   descripcion VARCHAR(255) NOT NULL,
   precio INT DEFAULT NULL,
   PRIMARY KEY (iduniverso)
);

CREATE TABLE IF NOT EXISTS universo_paralelo (
   iduniverso INT NOT NULL,
   descripcion VARCHAR(255) NOT NULL,
   precio INT DEFAULT NULL,
   PRIMARY KEY (iduniverso)
);


CREATE TABLE IF NOT EXISTS horarios (
   id_registro INT NOT NULL,
   id_usuario INT NOT NULL,
   fecha_reportada TIMESTAMP,
   PRIMARY KEY (id_registro)
);

CREATE TABLE IF NOT EXISTS datos (
   id_registro INT NOT NULL,
   id_usuario INT NOT NULL,
   fecha_reportada TIMESTAMP,
   PRIMARY KEY (id_registro)
);

CREATE TABLE IF NOT EXISTS info (
   id_info SERIAL,
   descripcion VARCHAR(255) NOT NULL,
   precio INT DEFAULT NULL,
   PRIMARY KEY (id_info)
);


CREATE TABLE IF NOT EXISTS participante (
  id INT NOT NULL,
  nombre VARCHAR(255) DEFAULT NULL,
  puntuacion FLOAT DEFAULT NULL
);

CREATE TABLE IF NOT EXISTS back_up (
   indice INT NOT NULL,
   id INT NOT NULL,
   user_id INT NOT NULL,
   marked_date TIMESTAMP NOT NULL,
   transaction_hash VARCHAR(255) NOT NULL,
   cifras INT NOT NULL,
   valores INT NOT NULL,
   email VARCHAR(255) NOT NULL,   
   uri TEXT NOT NULL,
   texto TEXT NOT NULL,
   ciudad VARCHAR(255) NOT NULL,     
   PRIMARY KEY (indice)
);

CREATE TABLE IF NOT EXISTS indices (
   indice INT NOT NULL,
   id INT NOT NULL,
   user_id INT NOT NULL,
   marked_date TIMESTAMP NOT NULL,
   transaction_hash VARCHAR(255) NOT NULL,
   cifras INT NOT NULL,
   valores INT NOT NULL,
   email VARCHAR(255) NOT NULL,   
   uri TEXT NOT NULL,
   texto TEXT NOT NULL,
   ciudad VARCHAR(255) NOT NULL,     
   PRIMARY KEY (indice)
);

CREATE TABLE IF NOT EXISTS inventario (
    id INT NOT NULL,
    seccion VARCHAR(255) NOT NULL,
    categoria VARCHAR(255) NOT NULL,
    cantidad INT NOT NULL
);

CREATE TABLE alimentos (
  idalimentos INT NOT NULL,
  alimento VARCHAR(45) NOT NULL,
  PRIMARY KEY (idalimentos)
);
  
CREATE TABLE bebidas (
  idbebidas int NOT NULL,
  bebida varchar(45) NOT NULL,
  precio int NOT NULL,
  PRIMARY KEY (idbebidas)
);

CREATE TABLE empleado (
  idEmpleado INT NOT NULL,
  nombre VARCHAR(45) NULL,
  pais VARCHAR(45) NULL,
  salario INT NULL,
  PRIMARY KEY (idEmpleado));
  
CREATE TABLE seccion (
  id INT NOT NULL,
  nombreSeccion VARCHAR(45) NOT NULL,
  idEmpleado INT NULL,
  cargo VARCHAR(45) NULL,
  PRIMARY KEY (id));
"""

postgres.execute(query)

sql_queries = query.replace('\n','').replace(';',';-').split('-')
sql_queries.remove('')

for query in sql_queries:
    mysql.execute(query)
```

# <center>INSERT</center>

```python
def insertar_en_universo():
    query = """
    INSERT INTO universo (iduniverso,descripcion,precio) 
    VALUES (7,'sol',23),(9,'luna',13),(78,'estrellas',25),(95,'galaxias',10),(1,'Loro',31),(2,'ToRo',78),(3,'molonico',2);
    """
    postgres.execute(query)
    mysql.execute(query) 
    
    print("Insercion Completada en UNIVERSO")

def listar_registros_universo():
    query = """
    SELECT * FROM universo
    """

    postgres_data = pd.read_sql(text(query), con=postgres)
    mysql_data = pd.read_sql(text(query), con=mysql)

    dataframe = pd.concat([postgres_data,mysql_data], axis=1)
    
    return dataframe

listar_registros_universo()
```

```python
def truncate_registros_universo_paralelo():
    query = """
    TRUNCATE TABLE universo_paralelo
    """
    postgres.execute(query)
    mysql.execute(query)
    
def insertar_en_universo_paralelo():
    
    truncate_registros_universo()
    insertar_en_universo()
    
    truncate_registros_universo_paralelo()
    
    query = """
    INSERT INTO universo_paralelo (iduniverso,descripcion,precio) 
        SELECT * FROM universo;
    """
    postgres.execute(query)
    mysql.execute(query) 
    
    query  = query.replace('\n','').replace('"','').replace('  ','').replace('\n','').replace('"','').replace('  ','')
    
    print(query)
    print("Insercion Completada en universo_paralelo")

def listar_registros_universo_paralelo():
    query = """
    SELECT * FROM universo_paralelo
    """

    postgres_data = pd.read_sql(text(query), con=postgres)
    mysql_data = pd.read_sql(text(query), con=mysql)

    dataframe = pd.concat([postgres_data,mysql_data], axis=1)
    
    return dataframe

truncate_registros_universo_paralelo()
listar_registros_universo_paralelo()
```

```python
query = """
SELECT *,
      ROW_NUMBER() OVER(PARTITION BY descripcion
      ORDER BY precio DESC) AS "# Sub Fila", 
      
      AVG(precio) OVER PARTICION_POR_DESCRIPCION AS Promedio,
      
      MIN(precio) OVER PARTICION_POR_DESCRIPCION AS Minimo,
      
      SUM(precio) OVER PARTICION_POR_DESCRIPCION AS Maximo,
      
      MIN(precio) OVER PARTICION_POR_DESCRIPCION AS Suma,
      
      COUNT(descripcion) OVER PARTICION_POR_CONTEO AS Total
FROM colores
    WHERE idcolores > 0
WINDOW
    PARTICION_POR_DESCRIPCION AS (PARTITION BY descripcion),
    PARTICION_POR_CONTEO AS (PARTITION BY descripcion)
ORDER BY descripcion
;
"""

postgres_data = pd.read_sql(text(query), con=postgres)
mysql_data = pd.read_sql(text(query), con=mysql)

pd.concat([postgres_data,mysql_data], axis=1)
```

```python
query = """
SELECT 
       ROW_NUMBER() OVER(PARTITION BY descripcion
       ORDER BY precio DESC) AS "Row Number", 
       precio,
       SUM(precio) OVER PARTICION_POR_DESCRIPCION_SUMA AS CumulativeTotal,
       descripcion
FROM colores
    WINDOW
            PARTICION_POR_DESCRIPCION_SUMA AS (PARTITION BY descripcion
                                           ORDER BY precio DESC 
                                           ROWS BETWEEN CURRENT ROW AND 1 FOLLOWING
                                          )
       ORDER BY descripcion, precio;
"""

postgres_data = pd.read_sql(text(query), con=postgres)
mysql_data = pd.read_sql(text(query), con=mysql)

pd.concat([postgres_data,mysql_data], axis=1)


       
```

# <center>RANKING</center>

## <p>Crea una numeración incremental de enteros empezando con el número 1 dentro de cada partición especificada. En esta numeración, se puede repetir la numeración si el ordenamiento dentro de la partición tiene el mismo valor, lo que se puede percibir como saltos de numeración.</p>

<center><img src="ranking.png"/></center>

```python
query = """ 
SELECT 
       ROW_NUMBER() OVER PARTICION_POR_DESCRIPCION_ROW_NUMBER AS "Row Number", 
       precio,
       RANK() OVER PARTICION_POR_RANKING AS RANKING,
       descripcion
FROM colores
    WINDOW
            PARTICION_POR_DESCRIPCION_ROW_NUMBER AS (PARTITION BY descripcion
                                                     ORDER BY precio DESC),
            PARTICION_POR_RANKING AS (PARTITION BY descripcion
                                      ORDER BY precio DESC 
                                      ROWS BETWEEN CURRENT ROW AND 1 FOLLOWING
                                     )
     ORDER BY descripcion, precio DESC, RANKING;
"""

postgres_data = pd.read_sql(text(query), con=postgres)
mysql_data = pd.read_sql(text(query), con=mysql)

pd.concat([postgres_data,mysql_data], axis=1)
```

```python
import json

def extraer_plan(tipo_query, query):
    
    if tipo_query.lower() == 'mysql':
        info_plan = query['EXPLAIN'][0]
        print(json.loads(info_plan))
    else:
        print(query['QUERY PLAN'][0][0])
        print('')
        
    
query_postgres = """
EXPLAIN  
    (FORMAT JSON)
SELECT * FROM numeros
    WHERE valor > 250 AND valor < 500
"""
postgres_data = pd.read_sql(query_postgres, con=postgres)

query_mysql = """
EXPLAIN  
    FORMAT=JSON
SELECT * FROM numeros
    WHERE valor > 250 AND valor < 500
"""
mysql_data = pd.read_sql(query_mysql, con=mysql)

pd.concat([postgres_data,mysql_data], axis=1)
```

## <center>INDICES DE TODAS LAS TABLAS</center>

```python
def indices_postgresql():
    
    query_postgres = """
    SELECT
        tablename,
        indexname,
        indexdef
    FROM
        pg_indexes
    WHERE
        schemaname = 'public'
    ORDER BY
        tablename,
        indexname;
    """    
    return pd.read_sql(query_postgres, con=postgres)

def indices_mysql():
    
    query_mysql = """
    SELECT DISTINCT
        TABLE_NAME,
        INDEX_NAME
    FROM INFORMATION_SCHEMA.STATISTICS
        WHERE TABLE_SCHEMA = 'digitos';
    """
    
    return pd.read_sql(query_mysql, con=mysql)


def listar_indices():
    mysql_data = indices_mysql()

    postgres_data =  indices_postgresql()

    return pd.concat([postgres_data,mysql_data], axis=1)

listar_indices()
```

## <center>INDICES DE TABLA ESPECIFICA</center>

```python
def listar_indices(tabla='numeros'):

    query_postgres = f"""
        SELECT
            indexdef,
            tablename,
            indexname
            
        FROM
            pg_indexes
        WHERE
            schemaname = 'public' AND tablename = '{tabla}'
        ORDER BY
            tablename,
            indexname;
    """

    postgres_data = pd.read_sql(query_postgres, con=postgres)

    query_mysql = f"""
        SELECT DISTINCT
            TABLE_NAME,
            INDEX_NAME,
            CARDINALITY
        FROM INFORMATION_SCHEMA.STATISTICS
        WHERE TABLE_NAME = '{tabla}'
        ORDER BY TABLE_NAME,
            INDEX_NAME;
    """
    mysql_data = pd.read_sql(text(query_mysql), con=mysql)
    
    return pd.concat([postgres_data,mysql_data], axis=1)
    

listar_indices(tabla='numeros')
```

```python
def listar_procedimientos_almacenados():
    query_mysql = """SHOW PROCEDURE STATUS;
    """

    query_postgres = """SELECT
        routine_schema,
        routine_name
    FROM 
        information_schema.routines
    WHERE 
        routine_type = 'PROCEDURE';
    """
    postgres_data = pd.read_sql(text(query_postgres), con=postgres)
    mysql_data = pd.read_sql(text(query_mysql), con=mysql)

    return pd.concat([postgres_data,mysql_data], axis=1)

listar_procedimientos_almacenados()
```

# <center> USER DEFINITION FUNCTIONS UDF</center>

```python
def listar_udf():
    query = """
    SELECT 
        n.nspname AS function_schema,
        p.proname AS function_name,
        l.lanname AS function_language,
        CASE WHEN l.lanname = 'internal' THEN p.prosrc
            ELSE pg_get_functiondef(p.oid)
            END AS definition,
        pg_get_function_arguments(p.oid) AS function_arguments,
        t.typname AS return_type
    FROM pg_proc p
        LEFT JOIN pg_namespace n ON p.pronamespace = n.oid
        LEFT JOIN pg_language l ON p.prolang = l.oid
        LEFT JOIN pg_type t ON t.oid = p.prorettype 
    WHERE
            l.lanname IN ('plpgsql') 
        AND 
            n.nspname NOT IN ('pg_catalog', 'information_schema')
    ORDER BY 
            function_language DESC, function_schema, function_name;

    """
    postgres_data = pd.read_sql(text(query), con=postgres)
    
    return pd.concat([postgres_data], axis=1)
```

```python
def listar_triggers():
    query_mysql = """
    SHOW TRIGGERS;
    """

    query_postgres = """
    SELECT  event_object_table AS table_name ,trigger_name         
        FROM information_schema.triggers  
    GROUP BY table_name , trigger_name 
        ORDER BY table_name ,trigger_name 
    """

    mysql_data = pd.read_sql(text(query_mysql), con=mysql)

    postgres_data = pd.read_sql(text(query_postgres), con=postgres)

    return pd.concat([postgres_data,mysql_data], axis=1)

listar_triggers()
```

# <center>CREAR TRIGGERS</center>

```python
query = """
CREATE OR REPLACE FUNCTION auditfuncInsert() 
RETURNS TRIGGER as
$$
   BEGIN
      INSERT INTO colores_auditoria(idcolores, precio, descripcion, accion, usuario, entry_date) 
        VALUES (NEW.idcolores, NEW.precio, NEW.descripcion,'Registro Insertado',current_user, current_timestamp);
      RETURN NEW;
   END;
$$
LANGUAGE plpgsql;

CREATE or replace TRIGGER log_insert
  AFTER INSERT
      ON colores
  FOR EACH ROW
      EXECUTE PROCEDURE auditfuncInsert();
  

CREATE OR REPLACE FUNCTION auditfuncDelete() 
RETURNS TRIGGER as
$$
   BEGIN
      INSERT INTO colores_auditoria(idcolores, precio, descripcion, accion, usuario, entry_date) 
        VALUES (OLD.idcolores, OLD.precio, OLD.descripcion,'Registro Eliminado',current_user, current_timestamp);
      RETURN NEW;
   END;
$$
LANGUAGE plpgsql;

CREATE or replace TRIGGER log_delete
  BEFORE DELETE
      ON colores
  FOR EACH ROW
      EXECUTE PROCEDURE auditfuncDelete();

CREATE OR REPLACE FUNCTION auditfuncUpdate() 
RETURNS TRIGGER as
$$
   BEGIN
      INSERT INTO colores_auditoria(idcolores, precio, descripcion, accion, usuario, entry_date) 
        VALUES (OLD.idcolores, OLD.precio, OLD.descripcion,'Registro Actualizado Valores Anteriores',current_user, current_timestamp),
            (NEW.idcolores, NEW.precio, NEW.descripcion,'Registro Actualizado Valores Nuevos',current_user, current_timestamp);
      RETURN NEW;
   END;
$$
LANGUAGE plpgsql;

CREATE or replace TRIGGER log_Update
  BEFORE UPDATE
      ON colores
  FOR EACH ROW
      EXECUTE PROCEDURE auditfuncUpdate(); 
"""
postgres.execute(query)
listar_triggers()
```

```python
query_mysql = """
SELECT seccion,
    SUM(IF(categoria='tenis', cantidad, NULL)) AS TENIS,
    AVG(IF(categoria='pantalon', cantidad, NULL)) AS PANTALONES,
    MAX(IF(categoria='camiseta', cantidad, NULL)) AS CAMISETAS,
    MIN(IF(categoria='chaqueta', cantidad, NULL)) AS CHAQUETAS
FROM inventario
    GROUP BY seccion
ORDER BY seccion;
"""
mysql_data = pd.read_sql(query_mysql, con=mysql)

pd.concat([mysql_data], axis=1)
```

# <center>UTILIDAD PARA GENERACION DE DATOS</center>

```python
import pandas as pd
import random 
from datetime import datetime, timedelta, date
import json
import string

def generador_limites_tiempo(iteracion):
    
    year = 2019 if iteracion%5 == 0 else 2020 if iteracion%4 == 0 else 2021 if iteracion%3 == 0 else 2022 
    month = random.randint(1,12)
    day = random.randint(1,31) if month in [1,3,5,7,8,10,12] else random.randint(1,30) if month in [4,6,9,11] else random.randint(1,28)
    user = random.randint(1,50)
    
    return year, month, day, user
    
    
def generador_lista_fechas_y_usuarios(limite:int = 100) -> list:
    
    listado = []
    
    for iteracion in range(limite):
        
        year, month, day, user = generador_limites_tiempo(iteracion)

        start = date(year, month, day)

        listado.append([str(user),str(start)])

    return listado


def eliminar_listas_duplicadas(lista_de_listas:list) -> list:
    
    return list(set([tuple(x) for x in lista_de_listas]))
    

def delta_time():    

    hours = random.randint(0,11)
    minutes = random.randint(0,59)
    seconds = random.randint(0,59)
    
    return hours, minutes, seconds

def generar_lista_para_insert(lista_validada:list) -> list():
    
    final_list = []
    counter = 1
    
    for indice, valor in enumerate(lista_validada):

        user = valor[0]
        fecha = valor[1].split('-')

        hours, minutes, seconds = delta_time()
        lower = timedelta(hours=hours, minutes=minutes, seconds=seconds)

        hours, minutes, seconds = delta_time()
        upper = timedelta(hours=hours, minutes=minutes, seconds=seconds)

        start = datetime(int(fecha[0]),int(fecha[1]),int(fecha[2])) + lower
        end = start + upper

        reg_1 = (counter, int(user), str(start))
        counter += 1
        
        reg_2 = (counter, int(user), str(end))
        counter += 1

        final_list.append(reg_1)
        final_list.append(reg_2)
        
    return final_list
        
def get_random_string(total_muestras):
    
    palabras = []
    
    for i in range(total_muestras):
    
        length = random.randint(20,50)
        letters = string.ascii_letters
        palabras.append(''.join(random.choice(letters) for i in range(length)))
        
    return palabras
    
    
def generar_cadena_query(data:str)->str:
    
    total_muestras = 50000
    
    lista_usuarios_fechas = generador_lista_fechas_y_usuarios(total_muestras)
    lista_validada = eliminar_listas_duplicadas(lista_usuarios_fechas)
    lista_para_insert = generar_lista_para_insert(lista_validada)
    
    if data == 'json':
        
        total_muestras = len(lista_para_insert)
        
        letras = get_random_string(total_muestras)
        
        values = pd.DataFrame(columns=['id','user_id','marked_date'], data=lista_para_insert)
        values['transaction_hash'] = letras
        values = values.reset_index().to_json(orient='records')
            
        json_file = json.loads(values)
       
                        
        with open('data_sql.json', 'w') as f:
            json.dump(json_file, f)
        
        return json_file
    
    return str(lista_para_insert).replace('[',' VALUES ').replace(']',';')
```

```python
generar_cadena_query('json')
```
