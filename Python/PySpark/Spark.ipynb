# <center> CARGA DATOS DE UN ARCHIVO JSON ALOJADO EN LA WEB

```python
import requests
import pandas as pd
tabla = 'https://raw.githubusercontent.com/JorgeCardona/skills/master/Python/DB/PeriodicTable.json'
info_tabla = requests.get(tabla).json().get('elements')
info_tabla  = pd.DataFrame(info_tabla)
info_tabla
```

# <center> CONVERTIR ARRAY DEL MISMO TIPO DE DATO

# <center> CONVERTIR ARRAY DEL MISMO TIPO DE DATO USANDO MAP, LAMBDA Y COMPREHENSION LIST

```python
listado  = list(map((lambda x: [float(i) if isinstance(i,(int,float)) else None for i in x]),info_tabla.ionization_energies.to_list()))
listado[0]
```

# <center> REASIGNAR VALORES TIPO FLOAT A LA COLUMNA ionization_energies

```python
info_tabla['ionization_energies'] = listado
info_tabla
```

# <center> GENERA UNA SESION PARA PROCESAR DATOS CON SPARK

```python
from pyspark.sql import SparkSession

spark = SparkSession \
                    .builder \
                    .appName("jorgeCardona") \
                    .master('local[*]') \
                    .getOrCreate()
spark
```

# <center> SPARK DATAFRAME DE LA TABLA PERIODICA

```python
df_tabla = spark.createDataFrame(info_tabla)
df_tabla.show(25,truncate=False), df_tabla.printSchema()
```

# <center> **HACER CONSULTA DIRECTAMENTE AL DATAFRAME Y CONVERTIRLO A PANDAS DATAFRAME** </center> 

```python
from pyspark.sql.functions import col
df_tabla.filter(col("name").like("%Gen%")).toPandas()
```

```python
df_tabla.filter(col("name").like("%gen%")).toPandas()
```

# <center> **HACER UNA CONSULTA CON SQL**

```python
tabla = 'tabla_periodica'
campos_seleccionar ='*'
campo_condicion ='name'

# crear una vista SQL temporal del dataframe
df_tabla.createOrReplaceTempView(tabla)

spark.sql("SELECT {campos_seleccionar} FROM {tabla} WHERE {campo_condicion} LIKE '%gen%' ".format(tabla=tabla,campos_seleccionar=campos_seleccionar, campo_condicion=campo_condicion)).toPandas()
```

```python
spark.sql("SELECT * FROM tabla_periodica WHERE phase LIKE '%gas%' ").toPandas()
```

```python
spark.sql("SELECT * FROM tabla_periodica WHERE phase LIKE '%Gas%' ").toPandas()
```

```python
spark.sql("SELECT * FROM tabla_periodica WHERE UPPER(phase) LIKE '%GAS%' ").toPandas()
```

```python
from pyspark.sql.functions import lower

df_tabla.filter(lower(col("phase")).like("%gas%")).toPandas()
```

# <center> FILTRAR CON NAN

```python
spark.sql("SELECT * FROM tabla_periodica WHERE UPPER(phase) LIKE '%GAS%' AND isNaN(molar_heat) = false").toPandas()
```

# <center> FILTAR CON NOT NULL

```python
spark.sql("SELECT * FROM tabla_periodica WHERE UPPER(phase) LIKE '%GAS%' AND named_by IS NOT NULL").toPandas()
```

```python
spark.sql("SELECT * FROM tabla_periodica \
            WHERE UPPER(phase) LIKE '%GAS%' AND named_by IS NOT NULL \
            AND size(ionization_energies) > 1").toPandas()
```

# <center> FUNCIONES DE AGREGACION

```python
spark.sql("SELECT SUM(period) AS PERIODO_TOTAL, MIN(size(ionization_energies)), MAX(size(ionization_energies)) , AVG(size(ionization_energies)) FROM tabla_periodica \
            WHERE UPPER(phase) LIKE '%GAS%' AND named_by IS NOT NULL").toPandas()
```
