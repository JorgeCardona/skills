# <center> GENERAR ARCHIVOS CSV Y EXCEL </center>

```python
archivo = 'archivo_referencia'
archivo_csv = archivo + '.csv'
archivo_excel = archivo + '.xlsx'
archivo_xml = archivo + '.xml'
archivo_json = archivo + '.json'
archivo_html = archivo + '.html'

def generar_archivos(dataframe):
    
    dataframe.to_csv(archivo_csv, na_rep='', index=False)
    dataframe.to_excel(archivo_excel, na_rep='', index=False)
    dataframe.to_xml(archivo_xml)
    dataframe.to_json(archivo_json)    
    dataframe.to_html(archivo_html, index=False)    
    
    
# guarda el dataframe en arhivos
generar_archivos(portapapeles)
```
