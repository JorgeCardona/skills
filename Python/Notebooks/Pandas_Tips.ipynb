# <center> GENERAR ARCHIVOS CSV Y EXCEL </center>

```python
archivo = 'archivo_referencia'
archivo_csv = archivo + '.csv'
archivo_excel = archivo + '.xlsx'

def generar_archivos(dataframe):
    
    dataframe.to_csv(archivo_csv, na_rep='', index=False)
    dataframe.to_excel(archivo_excel, na_rep='', index=False)
    
# guarda el dataframe en arhivos
generar_archivos(portapapeles)
```

# <center> LEER EXCEL  E IGNORAR ALGUNAS FILAS</center>

```python
pandas_excel = pd.read_excel(archivo_excel, skiprows=[0,1,2])
pandas_excel
```

```python
columna_comparar = 'Tres'
valor_comparar = 'i'
valor_reemplazo = 13
columna_a_cambiar = 'NUEVE'
pandas_excel.loc[pandas_excel[columna_comparar] == valor_comparar, columna_a_cambiar] = valor_reemplazo
pandas_excel
```
