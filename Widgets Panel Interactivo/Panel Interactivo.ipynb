### <center> GENERALIZACION O **REUSO DE UNA INTERFACE, CON UNA FUNCION AL GENERAR UN EVENTO** PARA CUALQUIER COMPONENTE QUE SE QUIERA USAR
## <center> **USANDO UNA FUNCION LAMBDA** PARA PASO DE PARAMETROS

# FUNCION Y CODIGO BASE

```python
import panel as pn

# create widgets
text = pn.widgets.StaticText()

# create function
def mostrar_valores(variable, parametro_1 = 0, parametro_2 =9):
    text.value = f"Las opciones {variable.value} se han seleccionado {len(variable.value)} elementos!, el parametro_1 es : {parametro_1} y el parametro_2 es : {parametro_2}"

# create button widget    
button = pn.widgets.Button(icon="warning", button_type='danger', name='WARNING')
```

# CHECKBOX GROUP

```python
variable_a_usar = checkbox_group
# layout, usa el widget de autocompletado, el widget del evento y el widget donde se muestra el contenido obtenido al procesar la funcion 
pn.Column(variable_a_usar, button, text)

# button with on click, llama a la funcion que interacciona al presionar el boton
button.on_click(lambda event: mostrar_valores(variable=variable_a_usar, parametro_1='Python', parametro_2='Java'))
pn.Column(variable_a_usar, button, text)
```

# MULTI SELECT

```python
variable_a_usar = multi_select
# layout, usa el widget de autocompletado, el widget del evento y el widget donde se muestra el contenido obtenido al procesar la funcion 
pn.Column(variable_a_usar, button, text)

# button with on click, llama a la funcion que interacciona al presionar el boton
button.on_click(lambda event: mostrar_valores(variable_a_usar))
pn.Column(variable_a_usar, button, text)
```

# CROSS SELECTOR

```python
variable_a_usar = cross_selector
# layout, usa el widget de autocompletado, el widget del evento y el widget donde se muestra el contenido obtenido al procesar la funcion 
pn.Column(variable_a_usar, button, text)

# button with on click, llama a la funcion que interacciona al presionar el boton
button.on_click(lambda event: mostrar_valores(variable_a_usar))
pn.Column(variable_a_usar, button, text)
```
